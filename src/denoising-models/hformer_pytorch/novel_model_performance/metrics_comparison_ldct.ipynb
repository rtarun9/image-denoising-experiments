```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt('ldct_denoising_metrics.txt')

# Extract columns
novel_psnr = data[:, 0]
noisy_psnr = data[:, 1]
noisy_ssim = data[:, 2]
novel_ssim = data[:, 3]
```

```python
novel_psnr.mean(), noisy_psnr.mean(), novel_ssim.mean(), noisy_ssim.mean()
```

```python
plt.figure(figsize=(20,10))
plt.plot(novel_psnr, 'r')
plt.plot(noisy_psnr)
```

```python
plt.figure(figsize=(20,10))
plt.plot(novel_ssim, 'r')
plt.plot(noisy_ssim)
```
