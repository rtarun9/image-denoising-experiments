```python
hecktor_dataset_dir = '../../../Dataset/HECKTOR/hecktor2022_training_corrected_v3/hecktor2022_training/hecktor2022/imagesTr/'
```

```python
# The dataset consist of nii images, which are 3D images (each image = a slice).
```

```python
import os
import numpy as np
import cv2 as cv
import tensorflow as tf
from glob import glob
import matplotlib.pylab as plt
import nibabel as nib
```

```python
hecktor_training_files = []

for _, _, file_names in os.walk(hecktor_dataset_dir):
    for file_name in file_names:
        hecktor_training_files.append(hecktor_dataset_dir + file_name)
```

```python
print('num of images: ', len(hecktor_training_files))
```

```python
img = nib.load(hecktor_training_files[0]).get_fdata()
print('shape of image: ', img.shape)
```

```python
# Displaying 16 slices.
plt.style.use('default')
fig, axes = plt.subplots(4,4, figsize=(24,24))
for i, ax in enumerate(axes.reshape(-1)):
    ax.set_title('slice: ' + str(i + 260))
    ax.imshow(img[:,:,360 + i])
plt.show()
```

```python
# Displaying 16 slices (in gray cmap).
plt.style.use('default')
fig, axes = plt.subplots(4,4, figsize=(24,24))
for i, ax in enumerate(axes.reshape(-1)):
    ax.set_title('slice: ' + str(i + 260))
    ax.imshow(img[:,:,360 + i], cmap='gray')
plt.show()
```
